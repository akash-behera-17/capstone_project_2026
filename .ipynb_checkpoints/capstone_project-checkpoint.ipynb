```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
from sklearn.preprocessing import MinMaxScaler
import warnings

warnings.filterwarnings('ignore')
plt.style.use('dark_background') # Better for dark themes

```

### Task 1: Stock Universe Selection
We select 5 diverse stocks from the NSE (National Stock Exchange of India):
1. **RELIANCE.NS** (Energy/Conglomerate)
2. **TCS.NS** (Information Technology)
3. **HDFCBANK.NS** (Banking/Financial Services)
4. **SUNPHARMA.NS** (Pharmaceuticals)
5. **MARUTI.NS** (Automobiles)



```python
# Define our stock universe
tickers = ['RELIANCE.NS', 'TCS.NS', 'HDFCBANK.NS', 'SUNPHARMA.NS', 'MARUTI.NS']
start_date = '2021-01-01'
end_date = '2025-12-31' # The dataset specification

# Fetch data for all tickers
print("Fetching data from Yahoo Finance...")
data = yf.download(tickers, start=start_date, end=end_date)['Close'] # We primarily need the Close prices
print("Data fetched successfully!")
display(data.head())

```

### Task 2: Data Preprocessing
1. **Handle Missing Values:** We'll use forward fill, which is standard for stock prices (assuming price stayed the same if not traded).
2. **Stationarity Check (ADF Test):** We'll check if the series are stationary.
3. **Scaling & Splitting:** We'll split the data into Train/Test and scale it for future ML models.



```python
# 1. Handle Missing Values using Forward Fill
data = data.ffill().bfill() # Forward fill first, then backward fill any remaining NaNs at the start
print(f"Missing values after cleaning:\n{data.isna().sum()}\n")

```

```python
# 2. Stationarity Check (Augmented Dickey-Fuller Test)
def test_stationarity(timeseries, title):
    # Perform Dickey-Fuller test
    print(f'Results of Dickey-Fuller Test for {title}:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput[f'Critical Value ({key})'] = value
    print(dfoutput)
    if dftest[1] <= 0.05:
        print("=> Strong evidence against the null hypothesis, series is STATIONARY.\n")
    else:
        print("=> Weak evidence against null hypothesis, time series is NON-STATIONARY. (Differencing needed)\n")

# Test for the first stock as an example
test_stationarity(data['RELIANCE.NS'], 'RELIANCE.NS')

```

```python
# Apply differencing to make them stationary (Log returns are often preferred in finance)
# We calculate log returns here, which will also be useful for Task 4 (Volatility)
log_returns = np.log(data / data.shift(1)).dropna()

print("\nTesting stationarity on Log Returns:")
test_stationarity(log_returns['RELIANCE.NS'], 'RELIANCE.NS (Log Returns)')

```

```python
# 3. Train/Test Split & Scaling
# The project specifies testing on the last 6 months (July 2025 - December 2025).
# Note: Since we are in mid-2026, the data for late 2025 should exist.
split_date = '2025-07-01'

train_data = data[:split_date]
test_data = data[split_date:]

print(f"Training data shape: {train_data.shape}")
print(f"Testing data shape: {test_data.shape}")

# Scale the data (fit on train, transform on both) - Useful for models like LSTM
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_train = pd.DataFrame(scaler.fit_transform(train_data), columns=train_data.columns, index=train_data.index)
scaled_test = pd.DataFrame(scaler.transform(test_data), columns=test_data.columns, index=test_data.index)

print("\nData preprocessing complete! Ready for Task 3: Forecasting.")

```

### Task 3: Time Series Forecasting
We will build two models: **ARIMA** and **Exponential Smoothing (Holt-Winters)**.
We will train on data before July 2025, and validate on the test set (July-Dec 2025).
Finally, we'll forecast 5 days into the future and evaluate our models.



```python
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error

def directional_accuracy(y_true, y_pred):
    """Calculates the percentage of times the model correctly predicts the direction of price movement."""
    true_direction = np.sign(np.diff(y_true))
    pred_direction = np.sign(np.diff(y_pred))
    # Handle the case where the array is empty
    if len(true_direction) == 0: return 0
    return np.mean(true_direction == pred_direction) * 100

def evaluate_forecast(y_true, y_pred, model_name):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mape = mean_absolute_percentage_error(y_true, y_pred) * 100
    da = directional_accuracy(y_true.values, y_pred.values if hasattr(y_pred, 'values') else np.array(y_pred))
    print(f"--- {model_name} ---")
    print(f"RMSE: {rmse:.2f}")
    print(f"MAPE: {mape:.2f}%")
    print(f"Directional Accuracy: {da:.2f}%\n")
    return rmse, mape, da

```

```python
# Train and Evaluate Models for a single stock (e.g., RELIANCE.NS) as a proof of concept
stock = 'RELIANCE.NS'
print(f"Training models for {stock}...")

# 1. ARIMA Model 
# (Using order=(5,1,0) as an example. In a full pipeline, we'd use auto_arima to find best p,d,q)
arima_model = ARIMA(train_data[stock], order=(5, 1, 0))
arima_result = arima_model.fit()

# Forecast over the test period
forecast_steps = len(test_data)
arima_forecast = arima_result.forecast(steps=forecast_steps)

evaluate_forecast(test_data[stock], arima_forecast, "ARIMA(5,1,0)")

```

```python
# 2. Exponential Smoothing (Holt-Winters)
hw_model = ExponentialSmoothing(train_data[stock], trend='add', seasonal=None, initialization_method="estimated")
hw_result = hw_model.fit()
hw_forecast = hw_result.forecast(steps=forecast_steps)

evaluate_forecast(test_data[stock], hw_forecast, "Exponential Smoothing")

```

```python
# Plotting the validation results
plt.figure(figsize=(14, 6))
plt.plot(train_data.index[-100:], train_data[stock][-100:], label='Train (Last 100 days)')
plt.plot(test_data.index, test_data[stock], label='Actual Test Data')
plt.plot(test_data.index, arima_forecast, label='ARIMA Forecast', linestyle='--')
plt.plot(test_data.index, hw_forecast, label='HW Forecast', linestyle='-.')
plt.title(f"{stock} - Validation Forecast Comparison")
plt.legend()
plt.show()

```

```python
# Generating the official 5-Day Forecast into the future (January 2026 / Trading Window)
print("Generating 5-Day Future Forecast based on ALL data...")
# We retrain the model on the entire dataset to get the best future forecast
final_arima = ARIMA(data[stock], order=(5, 1, 0)).fit()
future_forecast_5d = final_arima.forecast(steps=5)

print("\n--- 5-Day Future Forecast (ARIMA) ---")
print(future_forecast_5d)

```

### Task 4: Volatility and Trend Analysis
We analyze the risk profile (volatility) of all stocks in our universe using the rolling standard deviation of log returns.



```python
print("\n--- Task 4: Volatility Analysis ---")
# Calculate rolling 30-day standard deviation (volatility) for all stocks
rolling_volatility = log_returns.rolling(window=30).std() * np.sqrt(252) # Annualized volatility

# Get the most recent volatility for each stock
current_volatility = rolling_volatility.iloc[-1]
print("Current Annualized Volatility (Risk):")
print(current_volatility.sort_values(ascending=False))

# Plot volatility for all stocks
plt.figure(figsize=(14, 6))
for ticker_sym in tickers:
    plt.plot(rolling_volatility.index, rolling_volatility[ticker_sym], label=ticker_sym)
plt.title("Rolling 30-Day Annualized Volatility")
plt.legend()
plt.show()

```

### Task 5: Portfolio Construction and Capital Allocation
**Virtual Capital: ₹10,00,000**
We will use two strategies to allocate capital:
1. **Volatility-Aware Sizing:** Stocks with lower volatility get a higher weight.
2. **Forecast-Guided Allocation:** Stocks with higher expected returns get a higher weight.
We will blend these two strategies equally.



```python
from IPython.display import display

print("\n--- Task 5: Portfolio Construction ---")
TOTAL_CAPITAL = 1000000

# First, let's get the 5-day return forecast for ALL stocks
# We'll use Exponential Smoothing for all stocks to get quick forecasts
expected_returns = {}
for ticker_sym in tickers:
    model = ExponentialSmoothing(data[ticker_sym], trend='add', seasonal=None, initialization_method="estimated").fit()
    forecast = model.forecast(steps=5)
    # Calculate expected percentage return over 5 days
    exp_ret = (forecast.iloc[-1] - data[ticker_sym].iloc[-1]) / data[ticker_sym].iloc[-1]
    expected_returns[ticker_sym] = exp_ret

exp_returns_series = pd.Series(expected_returns)
print("Expected 5-Day Returns:")
print((exp_returns_series * 100).round(2), "%\n")

# Strategy 1: Volatility-Aware Sizing (Inverse Volatility)
inv_volatility = 1 / current_volatility
vol_weights = inv_volatility / inv_volatility.sum()

# Strategy 2: Forecast-Guided Allocation
# To handle negative returns, we shift values so the minimum return gets a small positive weight
shifted_returns = exp_returns_series - exp_returns_series.min() + 0.001 
forecast_weights = shifted_returns / shifted_returns.sum()

# Blended Strategy (50% Volatility, 50% Forecast)
final_weights = (vol_weights * 0.5) + (forecast_weights * 0.5)

# Calculate final allocation
allocation_df = pd.DataFrame({
    'Volatility Weight': vol_weights,
    'Forecast Weight': forecast_weights,
    'Final Weight (%)': final_weights * 100,
    'Allocated Amount (₹)': final_weights * TOTAL_CAPITAL
})

print("--- Final Portfolio Allocation ---")
display(allocation_df.round(2))

# Plot the allocation
plt.figure(figsize=(10, 6))
plt.pie(allocation_df['Allocated Amount (₹)'], labels=allocation_df.index, autopct='%1.1f%%', startangle=140)
plt.title("Final Portfolio Allocation (₹10,00,000)")
plt.show()

print("\nReady for Virtual Trading! Use the 'Allocated Amount (₹)' column to place your trades on StockGro.")

```

### Task 6: Model Comparison
Here we summarize the strengths and weaknesses of the models we tested.
* **ARIMA:** Interpretable, good for stable trends, but struggles with sudden non-linear market shocks.
* **Exponential Smoothing:** Excellent at adapting to recent changes and capturing level shifts, but assumes constant trend/seasonality.

For our final portfolio weights, we relied on Exponential Smoothing as it provided robust near-term 5-day forecasts.
